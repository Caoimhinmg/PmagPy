## Problem 1

The flipping condition is fulfilled when both of the following equations are zero:

$$
{dE \over {d\theta}} = 2 K_u \sin \theta \cos \theta - M_s B \sin (\phi - \theta),
$$

and again

$$
{d^2 E \over {d\theta^2}} = 2 K_u \cos (2\theta) + M_s B \cos (\phi - \theta ).
$$

For $\phi = 180, \theta = 0$ (originally antiparallel), we have:

$${d^2 E \over {d\theta^2}} = 2 K_u[1] + M_s B[-1] = 0$$
so, solving for $B$ and setting equal to $B_c$ by definition:
$$
B_c = { {2K_u}\over{M_s} }
$$

Because $B_c=\mu_o H_c, H_c = B_c/\mu_o$.

```python
from IPython.display import Image
Image(filename='ps5_prob2a.png')
```

The red loop shows the hysteresis behavior as measured.  The quasi-linear portion at high fields is the ``high field slope'' $\chi_{hf}$ which is the paramagnetic slope.  All the ferromagnetic contributions are at saturation (the loop is closed and reversible).  Subtracting off this high field slope gives the blue line, which allows calculation of the saturation magnetization ($M_s$), which is the intercept of the zero slope line at high fields with the $y$ axis.  The intercept of the blue line gives the magnetization remaining at zero field after exposure to a high saturating field, or the saturation remanence ($M_r$).  The intercept of the hysteresis loop with the $y$ axis is the coercive field $\mu_o H_c$ (or $B_c$ in the figure).  b) The blue line is the difference between descending and ascending (upper and lower respectively) curves in a).  This essentially represents the fraction of magnetic moments that have flipped at a given field - none at 0 and all after about .35 T. The field at which half the moments have flipped ($\Delta M = 0.5)$ is the coercivity of remenance, here 100 mT.  c) This curve is the derivative of the curve in b) and is essentially the distribution of switching fields.    There is a single distribution with a peak at about the coercivity of remanence.    This specimen is probably very nearly single domain (uniaxial) magnetite because the $M_r/M_s$ ratio is very close to the ideal of 0.5 and the coercivity is very close to the ideal maximum coercivity of magnetite.  There is a single phase responsible for the ferromagnetic behavior.

The "Day" plot (a) shows $M_r/M_s$ versus $H_{cr}/H_c$.  The former is also known as the ``squareness'' and is a strong function of grain size and domain state. Squareness for unixial single domain particles is 0.5 and a depression can be caused by many factors, such as flowering, vortex structures, domain walls and even superparamagnetism.  The ratio of coercivity of remanence to coercivity is also a function of domain state, mineralogy, state of stress and the ideal for uniaxial, single domain magnetite is close to 1.  The general trend from high squareness, low  $H_{cr}/H_c$ to low squareness/ high $H_{cr}/H_c$  is consistent with a trend in magnetic grain  size from near single domain to larger  particles with vortex structures at the larger end.  b) and c) These plots show the trends of squareness versus coercivity  and squarenes versus coercivity of remanence separately.  There is no clear trend with coercivity of remenance, but there is a trend with coercivity.  This suggests that the trend in squareness is not related to a change in something carrying the remanence, but perhaps the addition of superparamagnetic material instead (which does not contribute to the coercivity of remanence.
